Map DEG Solyc IDs to Entrez IDs

```python
library(biomaRt)
library(dplyr)

# Load DEG data
leaf_sig <- read.csv("/mnt/results/leaf_deseq2/leaf_Flight_vs_Ground_significant.csv", stringsAsFactors=FALSE, row.names=1)
root_sig <- read.csv("/mnt/results/root_deseq2/root_Flight_vs_Ground_significant.csv", stringsAsFactors=FALSE)

# Load comprehensive mapping
all_tomato_mapping <- readRDS("/mnt/shared-workspace/tomato_entrez_mapping.rds")

# Function to map Solyc IDs to Entrez IDs
# Try exact match first, then version-stripped match
map_to_entrez <- function(gene_ids, mapping_df) {
  # Exact match
  exact <- match(gene_ids, mapping_df$solyc_id)
  mapped_exact <- !is.na(exact)
  
  # For unmapped, try version-stripped match
  unmapped_ids <- gene_ids[!mapped_exact]
  unmapped_noversion <- sub("\\.[0-9]+$", "", unmapped_ids)
  
  # Create version-stripped lookup (take first Entrez ID if multiple)
  noversion_lookup <- mapping_df %>%
    filter(!is.na(entrezgene_id) & entrezgene_id != "") %>%
    group_by(solyc_noversion) %>%
    summarise(entrezgene_id = first(entrezgene_id), .groups="drop")
  
  noversion_match <- match(unmapped_noversion, noversion_lookup$solyc_noversion)
  mapped_noversion <- !is.na(noversion_match)
  
  # Combine results
  result <- data.frame(
    gene_id = gene_ids,
    entrez_id = NA_character_,
    gene_name = NA_character_,
    stringsAsFactors = FALSE
  )
  
  # Fill exact matches
  result$entrez_id[mapped_exact] <- as.character(mapping_df$entrezgene_id[exact[mapped_exact]])
  result$gene_name[mapped_exact] <- mapping_df$external_gene_name[exact[mapped_exact]]
  
  # Fill version-stripped matches for previously unmapped
  if (sum(mapped_noversion) > 0) {
    idx_unmapped <- which(!mapped_exact)
    idx_to_fill <- idx_unmapped[mapped_noversion]
    result$entrez_id[idx_to_fill] <- as.character(noversion_lookup$entrezgene_id[noversion_match[mapped_noversion]])
  }
  
  return(result)
}

# Map leaf DEGs
leaf_solyc <- leaf_sig$gene_id[grepl("^Solyc", leaf_sig$gene_id)]
leaf_map <- map_to_entrez(leaf_solyc, all_tomato_mapping)
cat("Leaf Solyc IDs:", length(leaf_solyc), "\n")
cat("Leaf mapped to Entrez:", sum(!is.na(leaf_map$entrez_id)), "\n")
cat("Leaf mapping rate:", round(sum(!is.na(leaf_map$entrez_id))/length(leaf_solyc)*100, 1), "%\n")

# Map root DEGs
root_solyc <- root_sig$gene_id[grepl("^Solyc", root_sig$gene_id)]
root_map <- map_to_entrez(root_solyc, all_tomato_mapping)
cat("\nRoot Solyc IDs:", length(root_solyc), "\n")
cat("Root mapped to Entrez:", sum(!is.na(root_map$entrez_id)), "\n")
cat("Root mapping rate:", round(sum(!is.na(root_map$entrez_id))/length(root_solyc)*100, 1), "%\n")

# Save mappings
saveRDS(leaf_map, "/mnt/shared-workspace/leaf_entrez_mapping.rds")
saveRDS(root_map, "/mnt/shared-workspace/root_entrez_mapping.rds")
```

Prepare 4 DEG gene sets with Entrez IDs

```python
# KEGG uses Entrez IDs directly. Our biomaRt mapping provides Entrez IDs.
# The ~57-60% mapping rate is acceptable for enrichment analysis.
# Let's now prepare the 4 gene sets and run enrichment.

library(clusterProfiler)
library(enrichplot)
library(ggplot2)

# Load DEG data
leaf_sig <- read.csv("/mnt/results/leaf_deseq2/leaf_Flight_vs_Ground_significant.csv", stringsAsFactors=FALSE, row.names=1)
root_sig <- read.csv("/mnt/results/root_deseq2/root_Flight_vs_Ground_significant.csv", stringsAsFactors=FALSE)

# Load mappings
leaf_map <- readRDS("/mnt/shared-workspace/leaf_entrez_mapping.rds")
root_map <- readRDS("/mnt/shared-workspace/root_entrez_mapping.rds")

# Split into up/down regulated in Flight vs Ground
# Leaf: log2FoldChange > 0 = up in Flight, < 0 = down in Flight
leaf_up_idx <- leaf_sig$log2FoldChange > 0
leaf_down_idx <- leaf_sig$log2FoldChange < 0

# Root: shrunk_log2FoldChange
root_up_idx <- root_sig$shrunk_log2FoldChange > 0
root_down_idx <- root_sig$shrunk_log2FoldChange < 0

cat("Leaf up in Flight:", sum(leaf_up_idx), "genes\n")
cat("Leaf down in Flight:", sum(leaf_down_idx), "genes\n")
cat("Root up in Flight:", sum(root_up_idx), "genes\n")
cat("Root down in Flight:", sum(root_down_idx), "genes\n")

# Get Entrez IDs for each gene set
# Leaf up
leaf_up_genes <- leaf_sig$gene_id[leaf_up_idx & grepl("^Solyc", leaf_sig$gene_id)]
leaf_up_entrez <- leaf_map$entrez_id[leaf_map$gene_id %in% leaf_up_genes]
leaf_up_entrez <- unique(na.omit(leaf_up_entrez))

# Leaf down
leaf_down_genes <- leaf_sig$gene_id[leaf_down_idx & grepl("^Solyc", leaf_sig$gene_id)]
leaf_down_entrez <- leaf_map$entrez_id[leaf_map$gene_id %in% leaf_down_genes]
leaf_down_entrez <- unique(na.omit(leaf_down_entrez))

# Root up
root_up_genes <- root_sig$gene_id[root_up_idx & grepl("^Solyc", root_sig$gene_id)]
root_up_entrez <- root_map$entrez_id[root_map$gene_id %in% root_up_genes]
root_up_entrez <- unique(na.omit(root_up_entrez))

# Root down
root_down_genes <- root_sig$gene_id[root_down_idx & grepl("^Solyc", root_sig$gene_id)]
root_down_entrez <- root_map$entrez_id[root_map$gene_id %in% root_down_genes]
root_down_entrez <- unique(na.omit(root_down_entrez))

cat("\nEntrez IDs per gene set:\n")
cat("Leaf up:", length(leaf_up_entrez), "\n")
cat("Leaf down:", length(leaf_down_entrez), "\n")
cat("Root up:", length(root_up_entrez), "\n")
cat("Root down:", length(root_down_entrez), "\n")

# Save gene sets
saveRDS(list(
  leaf_up = leaf_up_entrez,
  leaf_down = leaf_down_entrez,
  root_up = root_up_entrez,
  root_down = root_down_entrez
), "/mnt/shared-workspace/deg_entrez_sets.rds")
```

Build sample-to-cell-type mapping from SOFT metadata

```python
# Build comprehensive sample-to-cell-type mapping
# Parse all sample titles and their column IDs from the SOFT file

soft_lines <- readLines(gzfile("/mnt/shared-workspace/GSE149217_family.soft.gz"))

# Extract all sample titles
title_lines <- soft_lines[grepl("^!Sample_title", soft_lines)]
titles <- sub("^!Sample_title = ", "", title_lines)

# For each sample, find the column ID in its description block
# The pattern is: each ^SAMPLE block has title and descriptions
# Let me parse more carefully by finding sample blocks

sample_map <- data.frame(
  title = character(),
  col_id = character(),
  stringsAsFactors = FALSE
)

i <- 1
while (i <= length(soft_lines)) {
  if (grepl("^\\^SAMPLE", soft_lines[i])) {
    # Find title
    title <- ""
    col_id <- ""
    j <- i + 1
    while (j <= length(soft_lines) && !grepl("^\\^SAMPLE", soft_lines[j])) {
      if (grepl("^!Sample_title", soft_lines[j])) {
        title <- sub("^!Sample_title = ", "", soft_lines[j])
      }
      # Find the column ID (2-letter + 2-digit pattern)
      if (grepl("^!Sample_description = [A-Z][0-9]", soft_lines[j])) {
        col_id <- sub("^!Sample_description = ", "", soft_lines[j])
      }
      j <- j + 1
    }
    if (title != "" && col_id != "") {
      sample_map <- rbind(sample_map, data.frame(title=title, col_id=col_id, stringsAsFactors=FALSE))
    }
    i <- j
  } else {
    i <- i + 1
  }
}

cat("Total samples mapped:", nrow(sample_map), "\n")
print(head(sample_map, 20))

# Decode cell types from title
# Pattern: CELLTYPE + METHOD + PLATE
# EP=Epidermis, EXO=Exodermis, COR=Cortex, EN=Endodermis, 
# X35S=35S control, MZ=Meristematic Zone, WOX=WOX4/vascular,
# ACT=ACT2/cortex, V=Vascular, PH=Phloem, XY=Xylem, MCO=Meristematic Cortex
# TRAP=TRAP-seq, INT=INTACT, SH=SHR, SCR=SCR, RSS1=RSS1, CMZ=CMZ

decode_celltype <- function(title) {
  if (grepl("^EP", title)) return("Epidermis")
  if (grepl("^EXO", title)) return("Exodermis")
  if (grepl("^COR", title)) return("Cortex")
  if (grepl("^EN", title)) return("Endodermis")
  if (grepl("^X35S", title)) return("35S_control")
  if (grepl("^MZ", title)) return("Meristematic_Zone")
  if (grepl("^WOX", title)) return("Vascular_WOX4")
  if (grepl("^ACT", title)) return("Cortex_ACT2")
  if (grepl("^VTRAP", title)) return("Vascular")
  if (grepl("^PH", title)) return("Phloem")
  if (grepl("^XY", title)) return("Xylem")
  if (grepl("^MCO", title)) return("Meristematic_Cortex")
  if (grepl("^SH", title)) return("SHR")
  if (grepl("^SCR", title)) return("SCR")
  if (grepl("^RSS", title)) return("RSS1")
  if (grepl("^CMZ", title)) return("Cortex_MZ")
  return("Unknown")
}

sample_map$cell_type <- sapply(sample_map$title, decode_celltype)
cat("\nCell type distribution:\n")
print(table(sample_map$cell_type))
```

```python
library(limma)
library(edgeR)

# Identify cell-type-enriched genes using limma-voom
# Compare each cell type vs 35S control

root_trap <- root_atlas  # already loaded, 56 columns
col_celltypes <- sapply(colnames(root_trap), function(cid) {
  idx <- match(cid, sample_map$col_id)
  if (!is.na(idx)) sample_map$cell_type[idx] else "Unknown"
})

# Remove Unknown and 35S_control for now, keep cell types
cell_types <- c("Epidermis", "Exodermis", "Cortex", "Cortex_ACT2", "Endodermis",
                "Meristematic_Cortex", "Meristematic_Zone", "Phloem", "Vascular", 
                "Vascular_WOX4", "Xylem")

# Filter genes: keep genes with CPM > 1 in at least 3 samples
dge <- DGEList(counts = root_trap)
dge <- calcNormFactors(dge)
cpm_mat <- cpm(dge)
keep <- rowSums(cpm_mat > 1) >= 3
cat("Genes passing CPM filter:", sum(keep), "out of", nrow(root_trap), "\n")

dge <- dge[keep, , keep.lib.sizes=FALSE]
dge <- calcNormFactors(dge)

# Create design matrix
# For each cell type, compare vs 35S_control
cell_type_factor <- factor(col_celltypes, levels = c("35S_control", cell_types))
design <- model.matrix(~ 0 + cell_type_factor)
colnames(design) <- levels(cell_type_factor)

# Voom transformation
v <- voom(dge, design, plot=FALSE)

# Fit model
fit <- lmFit(v, design)

# For each cell type, find enriched genes vs 35S control
celltype_markers <- list()

for (ct in cell_types) {
  contrast_str <- paste0(ct, " - 35S_control")
  contrast_mat <- makeContrasts(contrasts = contrast_str, levels = design)
  fit2 <- contrasts.fit(fit, contrast_mat)
  fit2 <- eBayes(fit2)
  
  res <- topTable(fit2, coef = 1, number = Inf, adjust.method = "BH")
  # Cell-type-enriched: logFC > 1, adj.P.Val < 0.05
  enriched <- res[res$logFC > 1 & res$adj.P.Val < 0.05, ]
  
  cat(ct, ":", nrow(enriched), "enriched genes (logFC>1, padj<0.05)\n")
  
  if (nrow(enriched) > 0) {
    celltype_markers[[ct]] <- rownames(enriched)
  }
}

saveRDS(celltype_markers, "/mnt/shared-workspace/root_celltype_markers.rds")
cat("\nTotal cell types with markers:", length(celltype_markers), "\n")
cat("Total marker genes across all cell types:", sum(sapply(celltype_markers, length)), "\n")
```

```python
# The full matrix is too large for in-memory processing. 
# Let's use a smarter approach: subsample cells for clustering, then assign markers
# Or better: use the published cell type markers from the paper's supplementary data

# The Yue et al. 2024 paper (doi: 10.1111/pce.14906) identified these cell types:
# mesophyll, epidermis, guard cells, vascular, trichome (type I, V, VI, VII)
# Let's use known tomato leaf marker genes from literature

# Known tomato leaf cell type markers (from published studies):
leaf_markers <- list(
  Mesophyll = c("Solyc03g005760", "Solyc03g005770", "Solyc03g005780",  # CAB genes
                "Solyc01g005300", "Solyc07g047860", "Solyc10g006230",   # Rubisco/chloroplast
                "Solyc01g005290", "Solyc08g067320"),                     # Photosynthesis
  Epidermis = c("Solyc06g068990", "Solyc06g069000", "Solyc06g069010",   # Cutin/wax genes
                "Solyc09g091730", "Solyc01g005470"),                      # EPIDERMAL
  Guard_cells = c("Solyc01g009530", "Solyc06g009500", "Solyc04g011530"), # Stomatal
  Vascular = c("Solyc01g094950", "Solyc07g053620", "Solyc03g096770"),    # Vascular markers
  Trichome = c("Solyc01g005300", "Solyc08g080370", "Solyc01g005470",     # Trichome
               "Solyc06g008530", "Solyc08g078790"),                      # Glandular
  Phloem = c("Solyc03g096770", "Solyc06g054130", "Solyc01g094950"),     # Phloem-specific
  Xylem = c("Solyc07g053620", "Solyc02g090330", "Solyc08g075750"),      # Xylem-specific
  Bundle_sheath = c("Solyc10g006230", "Solyc03g005760")                  # Bundle sheath
)

# Actually, let me take a more systematic approach.
# Instead of guessing markers, let me use the scRNA-seq data but with a subsample
# to make it computationally feasible

library(Matrix)

# Load matrix
mat <- readMM(gzfile("/mnt/shared-workspace/GSE201931_matrix.mtx.gz"))
gene_ids <- read.delim(gzfile("/mnt/shared-workspace/GSE201931_features.tsv.gz"), header=FALSE, stringsAsFactors=FALSE)$V2
rownames(mat) <- gene_ids

# Subsample 5000 cells for faster processing
set.seed(42)
cell_idx <- sample(ncol(mat), min(5000, ncol(mat)))
mat_sub <- mat[, cell_idx]

# Filter genes: expressed in >= 10 cells in the subsample
genes_keep <- rowSums(mat_sub > 0) >= 10
mat_sub <- mat_sub[genes_keep, ]
cat("Subsampled matrix:", dim(mat_sub), "\n")

# Simple normalization: log1p(CPM)
lib_size <- colSums(mat_sub)
mat_cpm <- sweep(mat_sub, 2, lib_size, FUN="/") * 1e6
mat_log <- log1p(as.matrix(mat_cpm))  # Convert to dense for 5000 cells

cat("Log-normalized matrix:", dim(mat_log), "\n")

# Find variable genes
gene_var <- apply(mat_log, 1, var)
top_var <- order(gene_var, decreasing=TRUE)[1:1500]
mat_var <- mat_log[top_var, ]

# PCA
pca_res <- prcomp(t(mat_var), center=TRUE, scale.=FALSE, rank.=30)
cat("PCA done. Variance explained (top 5 PCs):", round(head(pca_res$sdev^2 / sum(pca_res$sdev^2), 5) * 100, 1), "%\n")

# K-means clustering
set.seed(42)
km <- kmeans(pca_res$x, centers=10, nstart=10)
cat("Cluster sizes:\n")
print(table(km$cluster))
```

Map leaf markers to gene names via biomaRt in batches

Fisher's exact test for leaf cluster-DEG enrichment

```python
import scanpy as sc
import numpy as np
import pandas as pd
from scipy import stats
import json

# Now let's do proper Fisher's exact test for enrichment of each cluster's markers in DEGs
# Load cross-reference data
with open("/mnt/shared-workspace/leaf_scrna_crossref.json", "r") as f:
    crossref = json.load(f)

# All genes in the scRNA-seq dataset (version-stripped)
all_scrna_genes = set(crossref['cluster_markers'][str(k)] for k in crossref['cluster_markers'])
# Flatten
all_marker_genes = set()
for k, v in crossref['cluster_markers'].items():
    all_marker_genes.update(v)

# Background: all genes in the scRNA-seq dataset
# Load the full gene list
with gzip.open("/mnt/shared-workspace/GSE201931_features.tsv.gz", 'rt') as f:
    scrna_all_genes = set(pd.read_csv(f, sep='\t', header=None)[1].apply(lambda x: x.rsplit('.', 1)[0]).values)

print(f"Background gene set: {len(scrna_all_genes)}")

# DEG sets
leaf_up = set(crossref['leaf_up'])
leaf_down = set(crossref['leaf_down'])
root_up = set(crossref['root_up'])
root_down = set(crossref['root_down'])

# Fisher's exact test for each cluster × DEG set
results = []
for cl in sorted(crossref['cluster_markers'].keys(), key=lambda x: int(x)):
    markers = set(crossref['cluster_markers'][cl])
    
    for deg_name, deg_set in [('leaf_up', leaf_up), ('leaf_down', leaf_down), 
                               ('root_up', root_up), ('root_down', root_down)]:
        # 2x2 contingency table
        overlap = markers & deg_set
        markers_only = markers - deg_set
        deg_only = deg_set - markers
        neither = scrna_all_genes - markers - deg_set
        
        a = len(overlap)  # in both marker and DEG
        b = len(markers_only)  # in marker but not DEG
        c = len(deg_only)  # in DEG but not marker
        d = len(neither)  # in neither
        
        # Fisher's exact test
        table = [[a, b], [c, d]]
        odds_ratio, p_value = stats.fisher_exact(table, alternative='greater')
        
        # Bonferroni correction later
        results.append({
            'cluster': int(cl),
            'deg_set': deg_name,
            'overlap': a,
            'markers_total': len(markers),
            'deg_total': len(deg_set),
            'odds_ratio': odds_ratio,
            'p_value': p_value
        })

results_df = pd.DataFrame(results)
results_df['p_adjust'] = results_df['p_value'] * len(results_df)  # Bonferroni
results_df['significant'] = results_df['p_adjust'] < 0.05

print("\n=== Fisher's Exact Test: Leaf scRNA-seq Cluster × DEG Enrichment ===")
print(f"Total tests: {len(results_df)}")
print(f"Significant (padj < 0.05): {sum(results_df['significant'])}")

# Show significant results
sig = results_df[results_df['significant']].sort_values('p_adjust')
if len(sig) > 0:
    print("\nSignificant enrichments:")
    print(sig.to_string(index=False))
else:
    print("\nNo significant enrichments after Bonferroni correction")
    # Show top results by p-value
    print("\nTop results (uncorrected):")
    top = results_df.nsmallest(20, 'p_value')
    print(top[['cluster', 'deg_set', 'overlap', 'markers_total', 'deg_total', 'odds_ratio', 'p_value']].to_string(index=False))

results_df.to_csv("/mnt/shared-workspace/leaf_cluster_deg_fisher_test.csv", index=False)
```

```python
import gzip
import pandas as pd
import numpy as np
from scipy import stats
from collections import Counter

# Load root atlas counts
root_atlas = pd.read_csv(gzfile("/mnt/shared-workspace/GSE149217_Tomato_ATLAS_raw_counts.csv.gz"), index_col=0)
print(f"Root atlas: {root_atlas.shape}")

# Build sample-to-cell-type mapping from SOFT file
soft_lines = []
with gzip.open("/mnt/shared-workspace/GSE149217_family.soft.gz", 'rt') as f:
    soft_lines = f.readlines()

sample_map = {}
current_title = None
for line in soft_lines:
    line = line.strip()
    if line.startswith('!Sample_title = '):
        current_title = line.replace('!Sample_title = ', '')
    elif line.startswith('!Sample_description = ') and current_title:
        desc = line.replace('!Sample_description = ', '')
        if len(desc) == 3 and desc[0].isalpha() and desc[1:].isdigit():
            sample_map[desc] = current_title
            current_title = None

print(f"Sample map entries: {len(sample_map)}")

def decode_celltype(title):
    if title.startswith('EP'): return 'Epidermis'
    if title.startswith('EXO'): return 'Exodermis'
    if title.startswith('COR'): return 'Cortex'
    if title.startswith('EN'): return 'Endodermis'
    if title.startswith('X35S'): return 'Control_35S'
    if title.startswith('MZ'): return 'Meristematic_Zone'
    if title.startswith('WOX'): return 'Vascular_WOX4'
    if title.startswith('ACT'): return 'Cortex_ACT2'
    if title.startswith('VTRAP'): return 'Vascular'
    if title.startswith('PH'): return 'Phloem'
    if title.startswith('XY'): return 'Xylem'
    if title.startswith('MCO'): return 'Meristematic_Cortex'
    if title.startswith('SH'): return 'SHR'
    if title.startswith('SCR'): return 'SCR'
    if title.startswith('RSS'): return 'RSS1'
    if title.startswith('CMZ'): return 'Cortex_MZ'
    return 'Unknown'

# Map column IDs to cell types
col_ids = root_atlas.columns.tolist()
col_celltypes = {}
for col_id in col_ids:
    if col_id in sample_map:
        col_celltypes[col_id] = decode_celltype(sample_map[col_id])
    else:
        col_celltypes[col_id] = 'Unknown'

ct_counts = Counter(col_celltypes.values())
print(f"\nCell type distribution:")
for ct, count in sorted(ct_counts.items()):
    print(f"  {ct}: {count}")

# Identify cell-type-enriched genes using simple fold-change vs 35S control
# For each cell type, compute mean CPM and compare to 35S control

# Normalize: CPM
lib_sizes = root_atlas.sum(axis=0)
root_cpm = root_atlas.div(lib_sizes, axis=1) * 1e6

# Get 35S control columns
control_cols = [c for c in col_ids if col_celltypes.get(c) == 'Control_35S']
control_mean = root_cpm[control_cols].mean(axis=1)

# For each cell type, find enriched genes (FC > 2, CPM > 5 in cell type)
cell_types = ['Epidermis', 'Exodermis', 'Cortex', 'Cortex_ACT2', 'Endodermis',
              'Meristematic_Cortex', 'Meristematic_Zone', 'Phloem', 'Vascular', 
              'Vascular_WOX4', 'Xylem']

root_markers = {}
for ct in cell_types:
    ct_cols = [c for c in col_ids if col_celltypes.get(c) == ct]
    if len(ct_cols) == 0:
        print(f"  {ct}: no columns found")
        continue
    
    ct_mean = root_cpm[ct_cols].mean(axis=1)
    fc = ct_mean / (control_mean + 0.1)  # Add pseudocount
    
    # Filter: FC > 2, CPM > 5 in cell type, expressed in >= 2 replicates
    enriched = (fc > 2) & (ct_mean > 5)
    
    # Also require expression in at least half the replicates
    min_reps = max(2, len(ct_cols) // 2)
    expressed_in = (root_cpm[ct_cols] > 1).sum(axis=1) >= min_reps
    enriched = enriched & expressed_in
    
    marker_genes = root_atlas.index[enriched].tolist()
    # Strip version suffixes for cross-referencing
    marker_genes_nov = [g.rsplit('.', 1)[0] for g in marker_genes]
    root_markers[ct] = marker_genes_nov
    
    print(f"  {ct}: {len(marker_genes)} enriched genes (FC>2, CPM>5)")

# Save root markers
import json
with open("/mnt/shared-workspace/root_celltype_markers_python.json", "w") as f:
    json.dump(root_markers, f)
```

Identify root cell-type-enriched genes from TRAP-seq

```python
import gzip
import pandas as pd
import numpy as np
from scipy import stats
from collections import Counter
import json

# Load root atlas counts
with gzip.open("/mnt/shared-workspace/GSE149217_Tomato_ATLAS_raw_counts.csv.gz", 'rt') as f:
    root_atlas = pd.read_csv(f, index_col=0)
print(f"Root atlas: {root_atlas.shape}")

# Build sample-to-cell-type mapping from SOFT file
with gzip.open("/mnt/shared-workspace/GSE149217_family.soft.gz", 'rt') as f:
    soft_lines = f.readlines()

sample_map = {}
current_title = None
for line in soft_lines:
    line = line.strip()
    if line.startswith('!Sample_title = '):
        current_title = line.replace('!Sample_title = ', '')
    elif line.startswith('!Sample_description = ') and current_title:
        desc = line.replace('!Sample_description = ', '')
        if len(desc) == 3 and desc[0].isalpha() and desc[1:].isdigit():
            sample_map[desc] = current_title
            current_title = None

print(f"Sample map entries: {len(sample_map)}")

def decode_celltype(title):
    if title.startswith('EP'): return 'Epidermis'
    if title.startswith('EXO'): return 'Exodermis'
    if title.startswith('COR'): return 'Cortex'
    if title.startswith('EN'): return 'Endodermis'
    if title.startswith('X35S'): return 'Control_35S'
    if title.startswith('MZ'): return 'Meristematic_Zone'
    if title.startswith('WOX'): return 'Vascular_WOX4'
    if title.startswith('ACT'): return 'Cortex_ACT2'
    if title.startswith('VTRAP'): return 'Vascular'
    if title.startswith('PH'): return 'Phloem'
    if title.startswith('XY'): return 'Xylem'
    if title.startswith('MCO'): return 'Meristematic_Cortex'
    if title.startswith('SH'): return 'SHR'
    if title.startswith('SCR'): return 'SCR'
    if title.startswith('RSS'): return 'RSS1'
    if title.startswith('CMZ'): return 'Cortex_MZ'
    return 'Unknown'

col_ids = root_atlas.columns.tolist()
col_celltypes = {}
for col_id in col_ids:
    if col_id in sample_map:
        col_celltypes[col_id] = decode_celltype(sample_map[col_id])
    else:
        col_celltypes[col_id] = 'Unknown'

ct_counts = Counter(col_celltypes.values())
print(f"\nCell type distribution:")
for ct, count in sorted(ct_counts.items()):
    print(f"  {ct}: {count}")

# Normalize: CPM
lib_sizes = root_atlas.sum(axis=0)
root_cpm = root_atlas.div(lib_sizes, axis=1) * 1e6

# Get 35S control columns
control_cols = [c for c in col_ids if col_celltypes.get(c) == 'Control_35S']
control_mean = root_cpm[control_cols].mean(axis=1)

# For each cell type, find enriched genes
cell_types = ['Epidermis', 'Exodermis', 'Cortex', 'Cortex_ACT2', 'Endodermis',
              'Meristematic_Cortex', 'Meristematic_Zone', 'Phloem', 'Vascular', 
              'Vascular_WOX4', 'Xylem']

root_markers = {}
for ct in cell_types:
    ct_cols = [c for c in col_ids if col_celltypes.get(c) == ct]
    if len(ct_cols) == 0:
        print(f"  {ct}: no columns found")
        continue
    
    ct_mean = root_cpm[ct_cols].mean(axis=1)
    fc = ct_mean / (control_mean + 0.1)
    
    enriched = (fc > 2) & (ct_mean > 5)
    min_reps = max(2, len(ct_cols) // 2)
    expressed_in = (root_cpm[ct_cols] > 1).sum(axis=1) >= min_reps
    enriched = enriched & expressed_in
    
    marker_genes = root_atlas.index[enriched].tolist()
    # Strip version suffixes (format: Solyc01g005000.3.1 -> Solyc01g005000)
    marker_genes_nov = [g.split('.')[0] for g in marker_genes]
    root_markers[ct] = marker_genes_nov
    
    print(f"  {ct}: {len(marker_genes)} enriched genes (FC>2, CPM>5)")

with open("/mnt/shared-workspace/root_celltype_markers_python.json", "w") as f:
    json.dump(root_markers, f)
print("\nSaved root cell type markers")
```

Fisher's exact test for root cell type-DEG enrichment

```python
import json
import pandas as pd
from scipy import stats

# Load root markers and DEGs
with open("/mnt/shared-workspace/root_celltype_markers_python.json", "r") as f:
    root_markers = json.load(f)

# Load DEGs
root_sig = pd.read_csv("/mnt/results/root_deseq2/root_Flight_vs_Ground_significant.csv")
leaf_sig = pd.read_csv("/mnt/results/leaf_deseq2/leaf_Flight_vs_Ground_significant.csv", index_col=0)

# Build DEG sets (version-stripped)
leaf_up = set(g.split('.')[0] for g in leaf_sig['gene_id'][(leaf_sig['log2FoldChange'] > 0) & leaf_sig['gene_id'].str.startswith('Solyc')])
leaf_down = set(g.split('.')[0] for g in leaf_sig['gene_id'][(leaf_sig['log2FoldChange'] < 0) & leaf_sig['gene_id'].str.startswith('Solyc')])
root_up = set(g.split('.')[0] for g in root_sig['gene_id'][(root_sig['shrunk_log2FoldChange'] > 0) & root_sig['gene_id'].str.startswith('Solyc')])
root_down = set(g.split('.')[0] for g in root_sig['gene_id'][(root_sig['shrunk_log2FoldChange'] < 0) & root_sig['gene_id'].str.startswith('Solyc')])

# Background: all genes in root atlas
with gzip.open("/mnt/shared-workspace/GSE149217_Tomato_ATLAS_raw_counts.csv.gz", 'rt') as f:
    first_line = f.readline()  # skip header
    bg_genes = set()
    for line in f:
        gene_id = line.split(',')[0].strip('"')
        bg_genes.add(gene_id.split('.')[0])

print(f"Background: {len(bg_genes)} genes")
print(f"DEG sets: leaf_up={len(leaf_up)}, leaf_down={len(leaf_down)}, root_up={len(root_up)}, root_down={len(root_down)}")

# Fisher's exact test for root cell types × DEG sets
results = []
for ct in sorted(root_markers.keys()):
    markers = set(root_markers[ct])
    
    for deg_name, deg_set in [('leaf_up', leaf_up), ('leaf_down', leaf_down), 
                               ('root_up', root_up), ('root_down', root_down)]:
        overlap = markers & deg_set
        markers_only = markers - deg_set
        deg_only = deg_set - markers
        neither = bg_genes - markers - deg_set
        
        a = len(overlap)
        b = len(markers_only)
        c = len(deg_only)
        d = len(neither)
        
        odds_ratio, p_value = stats.fisher_exact([[a, b], [c, d]], alternative='greater')
        
        results.append({
            'cell_type': ct,
            'deg_set': deg_name,
            'overlap': a,
            'markers_total': len(markers),
            'deg_total': len(deg_set),
            'pct_markers_in_deg': round(100*a/len(markers), 1) if len(markers) > 0 else 0,
            'pct_deg_in_markers': round(100*a/len(deg_set), 1) if len(deg_set) > 0 else 0,
            'odds_ratio': round(odds_ratio, 2),
            'p_value': p_value
        })

results_df = pd.DataFrame(results)
n_tests = len(results_df)
results_df['p_adjust'] = (results_df['p_value'] * n_tests).clip(upper=1.0)
results_df['significant'] = results_df['p_adjust'] < 0.05

print(f"\n=== Root Cell Type × DEG Fisher's Exact Test ===")
print(f"Total tests: {n_tests}")
print(f"Significant (Bonferroni padj < 0.05): {sum(results_df['significant'])}")

# Show all results sorted by p-value
print("\nAll results (sorted by p-value):")
top = results_df.sort_values('p_value')
print(top[['cell_type', 'deg_set', 'overlap', 'markers_total', 'deg_total', 
           'pct_markers_in_deg', 'odds_ratio', 'p_value', 'p_adjust', 'significant']].to_string(index=False))

results_df.to_csv("/mnt/shared-workspace/root_celltype_deg_fisher_test.csv", index=False)
```
